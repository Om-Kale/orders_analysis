```python
# we have "Not available", "unknown", nan this representing the null values nan is representing null
# so we convert Not Available, Unkonwn into null values or nan
df = pd.read_csv("orders.csv",na_values = ['Not Available', 'unknown'])
# na_values change it to null values 
```

```python
df.head(10)
```

```python
df['Ship Mode'].unique()
```

```python
# rename columns name : make it lowercase and remove space between 2 words and replace it with underscore _
df.columns
```

```python
df.columns=df.columns.str.lower()
```

```python
df.columns
```

```python
#replacing space by underscore _
df.columns = df.columns.str.replace(" ","_")
```

```python
df.head(5)
```

```python
# creating discount column 
df['discount']=df["list_price"]*df["discount_percent"]*0.01 # 0.01 = 100
# sale price
df['sale_price']=df['list_price'] - df['discount']
# profit
df["profit"] = df['sale_price'] - df['cost_price']
df.head(5)
```

```python
# check the data types 
df.dtypes
```

```python
# change the data type of order_date to datetime
#pd.to_datetime(df['order_date'],format = "%Y-%m-d")
df['order_date'] = pd.to_datetime(df['order_date'], format="%Y-%m-%d")
df.dtypes
```

```python
df.head(5)
```

```python
# drop unnecessary columns
df.drop(columns = ['cost_price'	,'list_price','discount_percent'],inplace = True)
```
